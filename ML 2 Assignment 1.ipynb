## Problem Statement
Build the linear regression model using scikit learn in boston data to predict 'Price' based on other dependent variable.

```python
#Importing packages
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston 
import scipy.stats as stats 
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
#load the boston data into the variable boston
boston = load_boston()
boston
```

```python
#checking the data variable of boston data
boston.data
```

```python
# independent variable in boston
boston.feature_names
```

```python
#Storing the boston data in dataframe
boston_data = pd.DataFrame(boston.data , columns=list(boston.feature_names))
boston_data.head()
```

```python
# checking the target variable of boston data 
boston.target
```

```python
#adding the target data to the boston dataframe
boston_data['target']=boston.target
boston_data.head()
```

```python
#describe the boston data set
boston_data.describe()
```

```python
#information about boston data set 
boston_data.info()
```

```python
# Conclusion from above information
# 1. all the independant variable is of type float and dependant variable is continuous variable
# 2. There is no missing values in the dataset 
```

```python
#seperate the independent variable from the dataset
x_data = boston_data.iloc[:,:13]
x_data.head()
```

```python
# Storing the target variable into y_data from the dataset
y_data = boston_data.iloc[:,13:14]
y_data.head()
```

```python
#split the data into train and test datsets
X_train, x_test ,y_train ,y_test = train_test_split(x_data,y_data)
```

```python
# Creating a linear regression object  
lm = LinearRegression()
lm
```

```python
#fit the model with train data 
model = lm.fit(X_train,y_train)
```

```python
#predict the model with the test data 
predict = model.predict(x_test)
predict
```

```python
# Model coefficients
model.coef_
```

```python
# coefficints of the Independent variables
pd.DataFrame(list(model.coef_),columns=list(x_data.columns))
```

```python
# Intercept of the linear regression line
model.intercept_
```

```python
# Calculating the rsquare score value
r2_values = r2_score(y_test,predict)
```

```python
# Calculating the mean square error value
mse = mean_squared_error(y_test,predict)
```

```python
print("R squared value : ", r2_values)
print("Mean Squared Error : ", mse)
```

The Linear Regression Model for the Boston dataset best fits the line with 74.99 % accuracy.
